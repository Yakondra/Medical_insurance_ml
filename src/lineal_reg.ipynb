# Model Linear regression

```python
import pandas as pd

X_train = pd.read_csv('../data/processed/medical-train_X_norm.csv')
X_test = pd.read_csv('../data/processed/medical-test_X_norm.csv')
y_train = pd.read_csv('../data/processed/ytrain.csv')
y_test = pd.read_csv('../data/processed/ytest.csv')

X_train.head()
```

```python
y_test.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()#, solver= 'newton-cholesky')
model.fit(X_train, y_train)
```

```python
print(f"Intercep (a): {model.intercept_}")
print(f"Coefficients (b1, b2): {model.coef_}")
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print(f"MSE: {mean_squared_error(y_test, y_pred)}")
print(f"R2 Score: {r2_score(y_test, y_pred)}")
```

```python
import math
rmse = math.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse}")
```
